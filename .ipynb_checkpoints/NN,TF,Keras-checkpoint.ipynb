### ACTIVATION FUNCTIONS

```python
##sigmoid
 
import math

def sigmoid(x):
    return 1/(1+math.exp(-x))
```

```python
sigmoid(-100)
```

```python
##tanh

def tanh(x):
    return ((math.exp(x)-math.exp(-x))/(math.exp(x)+math.exp(-x)))
```

```python
tanh(-100)
```

```python
#relu

def relu(x):
    return max(0,x)
```

```python
relu(-1)
```

```python
relu(2)
```

```python
##leaky relu

def leaky_relu(x):
    return max(0.1*x,x)
```

```python
leaky_relu(-1)
```

```python
leaky_relu(34)
```
